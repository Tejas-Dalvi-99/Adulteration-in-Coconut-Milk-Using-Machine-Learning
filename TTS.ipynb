```python
import pandas as pd
import numpy as np
```

```python
milk = pd.read_excel('/content/2. 2021_Raw Data.xlsx')
milk.describe()
```

```python
milk.info()
milk1= milk.drop(['Coconut_milk_type'], axis = 1)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
# Considering only numerical data
milk1.data = milk1.iloc[:, 1:]
```

```python
milk_std = scale(milk1.data)
milk_std
```

```python
pca = PCA(n_components = 2)
pca_values = pca.fit_transform(milk_std)
```

```python
# The amount of variance that each PCA explains is
var = pca.explained_variance_ratio_
var
```

```python
# PCA weights
pca.components_
pca.components_[0]
```

```python
# Cumulative variance
var1 = np.cumsum(np.round(var, decimals = 2) * 100)
var1
```

```python
# Variance plot for PCA components obtained
plt.plot(var1, color = "red")
```

```python

# PCA scores
pca_values
```

```python
df = pd.DataFrame(pca_values, columns = ['Column_A','Column_B'])

print(df)
```
